# Visualization

## 1. Cài đặt thư viện

```python
# 1. Import thư viện
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
from sklearn.decomposition import PCA
from sklearn.preprocessing import StandardScaler
from sklearn.cluster import KMeans
import pycountry
```

## 2. Chuẩn bị dữ liệu

```python
confirmed = pd.read_csv("../data/2_processed/01_confirmed_long.csv", parse_dates=["Date"])
deaths    = pd.read_csv("../data/2_processed/02_deaths_long.csv", parse_dates=["Date"])
recovered = pd.read_csv("../data/2_processed/03_recovered_long.csv", parse_dates=["Date"])
```

## 3. Tổng quan (Overview)

### 3.1 Top 10 quốc gia Confirmed, Deaths, Recovered

```python
# Tổng số theo quốc gia
total_confirmed = confirmed.groupby("Country/Region")["Confirmed"].max()
total_deaths = deaths.groupby("Country/Region")["Deaths"].max()
total_recovered = recovered.groupby("Country/Region")["Recovered"].max()

top10 = pd.DataFrame({
    "Confirmed": total_confirmed.sort_values(ascending=False).head(10),
    "Deaths": total_deaths.sort_values(ascending=False).head(10),
    "Recovered": total_recovered.sort_values(ascending=False).head(10)
})

top10
```

### 3.2 Map heatmap ca nhiễm theo quốc gia

```python
country_confirmed = confirmed.groupby("Country/Region")["Confirmed"].max().reset_index()
fig = px.choropleth(country_confirmed,
                    locations="Country/Region",
                    locationmode="country names",
                    color="Confirmed",
                    color_continuous_scale="Reds",
                    title="Tổng quan ca nhiễm theo quốc gia")
fig.show()
```
